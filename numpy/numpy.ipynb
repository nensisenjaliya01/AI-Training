```python
#[start:end:step]

import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr[0:4])
print(arr[4:])
print(arr[:4])
print(arr[-3:-1])
print(arr[0:5:2])
print(arr[0:5:1])
```

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr[0, 1:4])
```
