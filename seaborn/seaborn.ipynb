```python
from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(random.normal(size=1000), kind="kde")
plt.show()
```

```python
import seaborn as sns
x = random.binomial(n=3, p=0.5, size=50)
y = random.normal(loc=2, scale=3, size=100)
sns.displot(random.binomial(n=10, p=0.5, size=100))
sns.displot(random.normal(loc=1, scale=4, size=100))
print(x)
print(y)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

sns.get_dataset_names()
penguins= sns.load_dataset( 'penguins')
penguins.head()
```

```python
penguins['island'].value_counts()
sns.set_style("darkgrid")

sns.scatterplot(data=penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island", palette="Set1")
sns.despine()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


penguins = sns.load_dataset("penguins")


sns.stripplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue="island",
    dodge=True
)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


penguins = sns.load_dataset("penguins")

sns.swarmplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue="island",
    dodge=True
)
sns.despine()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


penguins = sns.load_dataset("penguins")


sns.histplot(
    data=penguins,
    x="species",
    hue = "sex",
    multiple= "stack"
)

plt.show()
```
