```python
class employee:
    def __init__(self, name, salary):
        self.name = name        # public
        self._age = 25          # protected
        self.__salary = salary  # private

    def get_salary(self):
        return self.__salary

    def set_salary(self, amount):
        if amount > 0:
            self.__salary = amount
e = employee("Nensi", 30000)

print(e.name)         
print(e._age) 
# print(e.__salary)
print(e.get_salary())
    
```
