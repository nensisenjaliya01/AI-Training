```python
numbers = [1, 2, 3]

for n in numbers:
    print(n)

it = iter(numbers) 
print(next(it))
print(next(it))
print(next(it))
```

Generator


```python
def number():
    yield 1
    yield 2
    yield 3

for n in numbers:
    print(n)
```

Decorator


```python
def hello():
    print("hello!!!!")

def my_decorator(func):
    def wrapper():
        print("before:")
        hello()
        print("after:")
    return wrapper
```

```python
def login_required(func):
    def wrapper():
        print("cheking login")
        func()
    return wrapper

@login_required
def dashbord():
    print("Welcome to dashbord")

dashbord()
```
