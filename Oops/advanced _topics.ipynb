Generator


```python
def numbers():
    for i in range(1,6):
        yield i
   

for n in numbers():
    print(n)
```

```python
def even(n):
    for n in range(2, n+1 ,2):
        yield n
for n in even(3):
    print(n)

```

```python
def squre():
    for i in range(1,11):
        yield i*i

for sq in squre():
    print(sq)
```

```python
def read_each_element(lst):
    for item in lst:
        yield item

data = [1, 2, 3]

for value in read_each_element(data):
    print(value)


```

```python
def vowels(text):
    for ch in text:
        if ch.lower() in "aeiou":
            yield ch

for v in vowels("nensi"):
    print(v)
```

```python
def read_file(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip

for line in read_file("document.txt"):
    print(line)
```

```python
import string
import random

def random_password(count, length=8):
        char = string.ascii_lowercase + string.digits
        for _ in range(count):
            yield ''.join(random.choice(char) for _ in range(length))

for psw in random_password(5):
    print(psw)
    
```

```python
from datetime import date, timedelta
def date_range(start, end):
    current = start
    while current <= end:
        yield current
        current += timedelta(days=1)

for d in date_range(date(2025,2,5), date(2025,2,7)):
    print(d)
    
```

```python
def fibonacci(n):
    a,b = 0,1
    for _ in range(n):
        yield a
        a, b=b, a+b
a = fibonacci(5)
print(list(a))
```

Decorator


```python
def hello():
    print("hello!!!!")

def my_decorator(func):
    def wrapper():
        print("before:")
        hello()
        print("after:")
    return wrapper
```

```python
import time

def monitor_speed(func):

    def wrapper(*arg, **kwargs):
        start = time.perf_counter()
        result = func(*arg, **kwargs)
        end = time.perf_counter()
        print(f" ---Execution time-{end-start:.6f}--time")
        return result
    return wrapper

        
@monitor_speed
def example_function():
    time.sleep(1)

example_function()

```

```python
def login_required(func):
    def wrapper():
        print("cheking login")
        func()
    return wrapper

@login_required
def dashbord():
    print("Welcome to dashbord")

dashbord()
```
