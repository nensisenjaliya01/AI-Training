```python
import time

def monitor_speed(func):

    def wrapper(*arg, **kwargs):
        start = time.perf_counter()
        result = func(*arg, **kwargs)
        end = time.perf_counter()
        print(f" ---Execution time-{end-start:.6f}--time")
        return result
    return wrapper

        
@monitor_speed
def example_function():
    time.sleep(1)

example_function()

```

```python
def login_required(func):
    def wrapper():
        print("cheking login")
        func()
    return wrapper

@login_required
def dashbord():
    print("Welcome to dashbord")

dashbord()
```
