GETTER/ SATTER METHOD


```python
class person:
    def __init__(self):
       self.__age = 20

    def get_age(self):
        return self.__age
    
    def set_age(self, value):
        if value > 0:
            self.__age = value
        else:
            print("age not a negative")
p = person()
print(p.get_age())
p.set_age(45)
print(p.get_age())

```

```python
class marks:
    def __init__(self):
        self.__marks = 67
    
    def get_marks(self):
        return self.__marks
    
    def set_marks(self, marks):
        if 0 <= marks >= 100:
            self.__marks = marks
        else:
            print("marks between 0 to 100:")
p = marks()
print(p.get_marks())
p.set_marks(100)
print(p.get_marks())
```

```python
class bankinfo:
    def __init__(self):
        self.__amount = 20000

    def get_amount(self):
        return self.__amount
    
    def set_amount(self, amount):
        if 0 <= amount:
            self.__amount = amount
        else:
            print("this is empty")
obj = bankinfo()
print(obj.get_amount())
obj.set_amount(9000)
print(obj.get_amount())

```

Overloading- same class, diffrent parameters


```python
class math:
    def cal(self, a, b=0, c=3):
        return a+b+c
obj = math()
print(obj.cal(1,1))
print(obj.cal(2,2))
```

Oerrriding-same class, same parameters

```python
class math:
    def A(self,a,b):
        return a+b
    def B(self,a,b):
        return a+b
obj = math()
print(obj.A(2,2))
print(obj.B(3,3))
```

MAP - apply funcation to each element

```python
a = [1,2,3]

result = map(lambda x: x*x, a)
print(list(result))
```

```python
names = ["nensi", "rahul"]

result = map(lambda x: x.upper(), names)

print(list(result))
```

Filter - select element based on condition

```python
a = [2,4,6]

result = filter(lambda x : x % 2 == 0, a)

print(list(result))

```

```python
marks = [10,50,60]

result = filter(lambda x: x >45, marks)

print("you are passed !!")
print(list(result))
```

REDUCE - it combine the elements

```python
from functools import reduce
 
numbers = [1, 2, 3, 4, 5, 6]

max_num = reduce(lambda x , y: x if x > y else y, numbers)

print(max_num)
```

```python
from functools import reduce

numbers = [1, 1, 1]

sum = reduce(lambda x , y: x + y, numbers )

print(sum)
```
