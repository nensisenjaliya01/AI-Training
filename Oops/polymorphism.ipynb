```python
class employee:
    def role(self):
        print("manager")
class person(employee):
    def role(self):
        print("CEO")
a = employee()
b = person()

a.role() #---same name but diffrent output----------------#
b.role()
```

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

a1 = Dog()
a2 = Cat()

a1.sound()
a2.sound()
```

```python
class Payment:
    def pay(self):
        print("Generic payment")

class UPI(Payment):
    def pay(self):
        print("Payment done using UPI")

class Card(Payment):
    def pay(self):
        print("Payment done using Debit/Credit Card")

class Cash(Payment):
    def pay(self):
        print("Payment done using Cash")

payments = [UPI(), Card(), Cash()]

for p in payments:
    p.pay()
```
