```python
class student:
    def __init__(self,name, age):
        self.name = name
        self.age = age

s1 = student("nensi", 20)
s2 = student("ruhu", 21)
print(s1.name, s1.age)
print(s2.name, s2.age)
```

```python
class employee:
    company_name = "Elision"
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
emp1 = employee("nensi",30000)
emp2 = employee("ruhu", 67000)
print(emp1.name, emp1.salary)
print(emp2.name, emp2.salary)
print("Before change:",emp1.company_name)

emp1.company_name = "cyber"

print("class variable company name(origin):",employee.company_name)
print("After change: ",emp1.company_name)
```
