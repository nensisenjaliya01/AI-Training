```python
# single inheritance
class animal:
    def speak(self):
        print("dog is speaking")
class dog(animal):
    def perk(self):
        print("dog is sitting")
d = dog()
d.speak()
d.perk()
        
```

```python
# multilevel inheritence
class animal:
    def speak(self):
        print("dog is speaking")
class dog(animal):
    def perk(self):
        print("dog is sitting")
class dogchild(dog):
    def cat(self):
        print("god is sleeping")
d = dogchild()
d.speak()
d.perk()
d.cat()
```

```python
# multiple inheritance
class calculate:
    def sum(self,a,b):
        return a + b
class multiplication:
    def mul(self,a,b):
        return a*b
class dirrived(calculate, multiplication):
    def div(self,a,b):
        return a/b
d = dirrived()
print(d.sum(1, 1))
print(d.mul(2, 2))
print(d.div(2, 4))
```

```python
#hirarcical inheritance
class parent:
    def fun1(self):
        print("this is 1st")
class child1(parent):
    def fun2(self):
        print("this is 2st")
class child2(parent):
    def fun3(self):
        print("this is 3st")
ob = child1()
ob2 = child2()

ob.fun1()
ob.fun2()
ob2.fun1()
ob2.fun3()


```

```python
class a:
    def fun1(self):
        print("1st")
class b(a):
    def fun2(self):
        print("2nd")
class c(a):
    def fun3(self):
        print("3rd")
class d(b, c):
    def fun4(self):
        print("4th")
ob1 = d()
ob1.fun1()
ob1.fun2()
ob1.fun3()
ob1.fun4()
print(issubclass(d,c))
print(issubclass(c,d))

        
```
