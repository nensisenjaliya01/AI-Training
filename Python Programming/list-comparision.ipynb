```python
number = [1, 2, 3, 4, -3]
number.sort()
print("ascending order:", number)
number.sort(reverse=True)
print("descending order:", number)
```

```python
number = [5, 3, 8, 6, 7]
n = len(number)

for i in range(n):
    for j in range(0, n - i - 1):
        if number[j] > number[j+1]:
            number[j], number[j+1] = number[j+1], number[j]
print("sorthed list:", number)
```
