```python
n = int(input("Enter num:"))
rev = 0
#reverse logic
for _ in range(len(str(n))):
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10
print(rev)
```

```python
#paligdrom
n = int(input("Enter number:"))
temp=n
rev=0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10

if temp == rev:
    print("it is paligdrom", rev, temp)
else:
    print("it is not")
```
