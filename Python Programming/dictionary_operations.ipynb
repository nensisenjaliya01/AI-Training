```python
s = "nensi"
freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
print(freq)
```

```python
data = {
    "emp1" : {"name":"nensi", "age":20, "salary":20},
    "emp2" : {"color":"red", "salary":20000 }
}
print(data["emp1"]["name"])
print(data["emp1"]["age"])
data["emp1"]["department"] = "AI"
print(data)
data["emp2"].pop("salary")
print(data)
count_emp = len(data)
print(count_emp)
count = 0
for emp in data.values():
    if "salary" in emp:
        count += 1
print(count)
for emp in data.values():
    if "name" in emp:
        print(emp["name"])
for emp in data.values():
    if "age" in emp and emp["age"] <= 21:
        print("you are eligible for voting")
```

```python
data = {
    "emp2": {"name": "rahul", "salary": 20000},
    "emp3": {"name": "priya", "salary": 60000}
}
for emp in data.values():
    if "salary" in emp:
        emp["salary"] = emp["salary"] * 0.10
print(data)
```
