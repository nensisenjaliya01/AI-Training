```python
s = "nensi"
freq = {}

for ch in s:
    if ch in freq:
        freq[s] += 1
    else:
        freq[s] = 1
print(freq)
```

```python
data = {"name": "nensi",
         "age": 20, 
         "color":None}
cleaned = {}
for key, value in data.items():
    if value is not None:
        cleaned[key] = value

print(cleaned)
```

```python
data = {"name": "nensi",
         "age": 20, 
         "color":None}
sorted_data = dict(sorted(data.items()))
print(sorted_data)
```

```python
data = {"a": 1, "b":4}
inverted = {}

for key, value in data.items():
    inverted[value] = key
print(inverted)
```
