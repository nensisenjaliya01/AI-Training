```python
sen = "Hello nensi welcome to python programming"
words = sen.split()
print(len(words))

longest = max(words, key=len)
print("longest word is:", longest)

uper = 0
lower = 0
for ch in sen:
    if ch.isupper():
        uper += 1
    elif ch.islower():
        lower += 1
print("uper:", uper)
print("lower:", lower)


freq={}
for ch in sen:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
print(freq)

print(sen.title())
print(sen.capitalize())
print(sen.swapcase())
print(sen.replace("nensi", "nancy"))
```
