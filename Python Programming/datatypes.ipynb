```python
def testing(user_input):
    valid_inputs = ["yes", "no", "maybe"]

    if (user_input in valid_inputs) and (user_input != ""):
        return True
    elif (user_input not in valid_inputs) or (user_input == ""):
        return False
    else:
        return None
    
```
